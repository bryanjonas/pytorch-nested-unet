```python
from skimage.io import imread
from glob import glob
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
```

```python
train_path = '/data/spacenet/bldg/data/pytorch-PAN/images/PAN_AOI_4_Shanghai_img5002.tif'
img = imread(train_path)
img = img.flatten()/img.max()
img = (img - img.mean()) / img.std()
sns.displot(img, stat="density")
```

```python
img2 = imread('WPTest.PNG')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img2 = img2.flatten()/img2.max()
img2 = (img2 - img2.mean()) / img2.std()
sns.displot(img2, stat="density")
```
